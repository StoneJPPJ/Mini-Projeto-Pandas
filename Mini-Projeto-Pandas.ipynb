### Importação modulos e arquivos com visualização.

```python
import pandas as pd
import numpy as np

# ler todos os csv's

baseServicos = pd.read_excel(r'BaseServiçosPrestados.xlsx')
cadastroCliente = pd.read_csv(r'CadastroClientes.csv', sep=';')
cadastroFuncionarios = pd.read_csv(r'CadastroFuncionarios.csv', sep=';', decimal=',')
```

```python
display(baseServicos, cadastroCliente, cadastroFuncionarios)
```

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa?

```python
# total da folha salarial
total_salario = cadastroFuncionarios['Salario Base'].sum()
total_beneficios = cadastroFuncionarios['Beneficios'].sum()
total_vt = cadastroFuncionarios['VT'].sum()
total_vr = cadastroFuncionarios['VR'].sum()
total_impostos = cadastroFuncionarios['Impostos'].sum()
total_folha_salarial = total_salario + total_beneficios + total_vt + total_vr + total_impostos
print(f'O total gasto pela empresa na folha salarial é de R${total_folha_salarial:,.2f}')
```

2. Qual foi o faturamento da empresa?<br>

```python
# juntar as bases de dados
faturamento_empresa = cadastroCliente.merge(baseServicos[['ID Cliente', 'Tempo Total de Contrato (Meses)']], on='ID Cliente')

# somar valor total das colunas 'Valor Contrato Mensal' e 'Tempo Total de Contratos(Meses)', depois apenas multiplica-lás
coluna_valor_contrato = faturamento_empresa['Valor Contrato Mensal'].sum()
coluna_tempo_contrato = faturamento_empresa['Tempo Total de Contrato (Meses)'].sum()
faturamento_total_empresa = coluna_tempo_contrato * coluna_valor_contrato

print(f'O faturamento total da empresa foi de R${faturamento_total_empresa:,.2f}')
```

3. Qual o % de funcionários que já fechou algum contrato?

```python
# quantidade de funcionarios
qtd_funcionarios = len(cadastroFuncionarios['ID Funcionário'])
contratos_funcionarios = baseServicos['ID Funcionário'].unique()
total_funcionarios_contratos = len(contratos_funcionarios)
print(f'A porcentagem de funcionarios que já fecharam algum contrato é de {total_funcionarios_contratos / qtd_funcionarios:.0%}.')
```

5. Calcule o total de funcionários por área


```python
def funcionarios_area(area, armazena):
    for contrato in cadastroFuncionarios['Area']:
        if contrato == area:
            armazena += 1
        else:
            pass
    return armazena
```
